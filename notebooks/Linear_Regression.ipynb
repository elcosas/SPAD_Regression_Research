```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

bounds = np.load("../proprietary/edh_boundaries.npy")
transients = np.load("../proprietary/transients_interp.npy")

tr1 = np.array_split(transients[0, :], 31)
for i, arr in enumerate(tr1):
    tr1[i] = np.mean(arr)
tr1 = np.array(tr1)
bd1 = np.array(range(0, np.max(bounds[0, :]), np.max(bounds[0, :]) // 31))[1:]
tr1.shape, bd1.shape
```

```python
lm = LinearRegression().fit(bd1.reshape(-1, 1), tr1)
pred_vals = lm.predict(bd1.reshape(-1, 1))
plt.scatter(bd1, tr1)
plt.plot(bd1, pred_vals, 'r')
```

```python
residuals = tr1 - pred_vals
fig, ax = plt.subplots()
ax.set_ylim(top=np.max(residuals), bottom=-np.max(residuals))
ax.scatter(bd1, residuals)
```
